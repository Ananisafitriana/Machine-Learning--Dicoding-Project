Tidak ada data kosong atau data duplikat pada dataset tersebut

````python
# Visualisasi boxplot untuk mendeteksi outlier 
plt.figure(figsize=(12, 6))

for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=bank[col])
    plt.title(f"Boxplot {col}")

plt.tight_layout()
plt.show()
````

````python
# Fungsi untuk menghapus outlier berdasarkan IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

initial_count = bank.shape[0]
columns_with_outliers = ["TransactionAmount", "TransactionDuration", "AccountBalance"]
for col in columns_with_outliers:
    bank = remove_outliers_iqr(bank, col)
final_count = bank.shape[0]

initial_count, final_count
````

````python
bank_encoded = pd.get_dummies(bank, columns=["TransactionType", "Channel"], drop_first=True)
````

````python
bank_encoded["TransactionAmount_log"] = np.log1p(bank_encoded["TransactionAmount"])
bank_encoded["TransactionDuration_log"] = np.log1p(bank_encoded["TransactionDuration"])
bank_encoded["AccountBalance_log"] = np.log1p(bank_encoded["AccountBalance"])

bank_encoded[["TransactionAmount_log", "TransactionDuration_log", "AccountBalance_log"]].describe()
````

````python
#Scaling
columns_to_scale = ["TransactionAmount_log", "TransactionDuration_log", "AccountBalance_log"]
scaler = StandardScaler()
bank_encoded[columns_to_scale] = scaler.fit_transform(bank_encoded[columns_to_scale])
````

````python
#binned 
bank_encoded['TransactionAmount_Binned'] = pd.qcut(bank_encoded['TransactionAmount'], q=3, labels=['Low', 'Medium', 'High'])
bank_encoded['CustomerAge_Binned'] = pd.cut(bank_encoded['CustomerAge'], bins=[17, 30, 45, 60, 100], labels=['Youth', 'Adult', 'Senior', 'Elder'])
````

````python
le_amount_bin = LabelEncoder()
le_age_bin = LabelEncoder()
bank_encoded['TransactionAmount_Binned'] = le_amount_bin.fit_transform(bank_encoded['TransactionAmount_Binned'])
bank_encoded['CustomerAge_Binned'] = le_age_bin.fit_transform(bank_encoded['CustomerAge_Binned'])
````

# **6. Pembangunan Model Clustering**

## **a. Pembangunan Model Clustering**

Pada tahap ini, Anda membangun model clustering dengan memilih algoritma yang sesuai untuk mengelompokkan data berdasarkan kesamaan. Berikut adalah **rekomendasi** tahapannya.
1. Pilih algoritma clustering yang sesuai.
2. Latih model dengan data menggunakan algoritma tersebut.

````python
scaled_data = bank_encoded[columns_to_scale].values 
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(scaled_data)
labels = kmeans.labels_
````

Note by Annisa : Saya menggunakan algoritma K-Means dalam pembuatan model clustering, untuk penggunaanya ada pada kode setelahnya

## **b. Evaluasi Model Clustering**

Untuk menentukan jumlah cluster yang optimal dalam model clustering, Anda dapat menggunakan metode Elbow atau Silhouette Score.

Metode ini membantu kita menemukan jumlah cluster yang memberikan pemisahan terbaik antar kelompok data, sehingga model yang dibangun dapat lebih efektif. Berikut adalah **rekomendasi** tahapannya.
1. Gunakan Silhouette Score dan Elbow Method untuk menentukan jumlah cluster optimal.
2. Hitung Silhouette Score sebagai ukuran kualitas cluster.

````python
# Mencari jumlah cluster dengan metode Elbow
wcss = [] 
K_range = range(1, 11) 
features_for_clustering = ["TransactionAmount_log", "TransactionDuration_log", "AccountBalance_log"]

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(bank_encoded[features_for_clustering])
    wcss.append(kmeans.inertia_) 

# Visualisasi Metode Elbow
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker="o", linestyle="-", color="b")
plt.xlabel("Jumlah Cluster (k)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Metode Elbow untuk Menentukan Jumlah Cluster Optimal")
plt.show()
````

````python
for k in range(2, 5):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_data)
    score = silhouette_score(scaled_data, labels)
    print(f'K={k}  Silhouette Score: {score:.3f}')
````

## **d. Visualisasi Hasil Clustering**

Setelah model clustering dilatih dan jumlah cluster optimal ditentukan, langkah selanjutnya adalah menampilkan hasil clustering melalui visualisasi.

Berikut adalah **rekomendasi** tahapannya.
1. Tampilkan hasil clustering dalam bentuk visualisasi, seperti grafik scatter plot atau 2D PCA projection.

````python
# Menentukan jumlah cluster yang dipilih
optimal_clusters = 4

kmeans_model = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
bank_encoded["Cluster"] = kmeans_model.fit_predict(bank_encoded[features_for_clustering])

plt.figure(figsize=(8, 6))
plt.scatter(bank_encoded["TransactionAmount_log"], bank_encoded["AccountBalance_log"], 
            c=bank_encoded["Cluster"], cmap="viridis", alpha=0.6, edgecolors="k")

centroids = kmeans_model.cluster_centers_

plt.scatter(centroids[:, features_for_clustering.index("TransactionAmount_log")], 
            centroids[:, features_for_clustering.index("AccountBalance_log")], 
            c="red", marker="X", s=200, label="Centroids")

plt.xlabel("TransactionAmount_log")
plt.ylabel("AccountBalance_log")
plt.title(f"Visualisasi Clustering dengan {optimal_clusters} Cluster")
plt.legend()
plt.show()
````

````python
# Inverse transform hasil encoding ke bentuk kategori aslinya
bank_encoded['TransactionAmount_Binned'] = le_amount_bin.inverse_transform(bank_encoded['TransactionAmount_Binned'])
bank_encoded['CustomerAge_Binned'] = le_age_bin.inverse_transform(bank_encoded['CustomerAge_Binned'])
````

Setelah melakukan clustering, langkah selanjutnya adalah menganalisis karakteristik dari masing-masing cluster berdasarkan fitur yang tersedia.

Berikut adalah **rekomendasi** tahapannya.
1. Analisis karakteristik tiap cluster berdasarkan fitur yang tersedia (misalnya, distribusi nilai dalam cluster).

````python


scaled_features = scaler.fit_transform(bank_encoded[features_for_clustering])

# Fit ulang scaler ke fitur clustering agar bisa di-inverse
original_values = scaler.inverse_transform(bank_encoded[features_for_clustering])
inverse_df = pd.DataFrame(original_values, columns=features_for_clustering)
inverse_df['Cluster'] = bank_encoded['Cluster'].values

# Analisis deskriptif berdasarkan hasil clustering
for cluster in sorted(inverse_df['Cluster'].unique()):
    print(f"Distribusi untuk Cluster {cluster}:")
    display(inverse_df[inverse_df['Cluster'] == cluster].describe())
    print("\n")


plt.figure(figsize=(8,5))
sns.countplot(data=inverse_df, x='Cluster', palette='viridis')
plt.title("Distribusi Jumlah Anggota per Cluster")
plt.xlabel("Cluster")
plt.ylabel("Jumlah Anggota")
plt.show()
````
